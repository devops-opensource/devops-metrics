**Configuration** 

Define configuration to be able to connect to a JIRA Cloud project:
* url
* project
* email
* apitoken (https://id.atlassian.com/manage-profile/security/api-tokens)

```python
config = {}
config["JIRA_CLOUD"] = {}
config["JIRA_CLOUD"]["jira_cloud_url"] = "https://daosio.atlassian.net/"
config["JIRA_CLOUD"]["jira_project_key"] = "CART"
config["JIRA_CLOUD"]["jira_user_email"] = ""
config["JIRA_CLOUD"]["jira_token"] = ""
config["JIRA_CLOUD"]["jira_creation_status"] = "Open"
config["JIRA_CLOUD"]["jira_released_status"] = "Released"
config["JIRA_CLOUD"]["jira_closed_statuses"] = "Closed"
```
